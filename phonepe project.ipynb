```python
#aggregate_transaction.

path1 = "C:/Users/mahes/OneDrive/Desktop/phonepe/pulse/data/aggregated/transaction/country/india/state/"
aggre_trans_state_list = os.listdir(path1)

columns1 = {"States":[], "Year":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[]}
for state_name in aggre_trans_state_list:
    curr_st_name = path1+state_name+"/"
    aggre_year_list = os.listdir(curr_st_name)
    
    for year_name in aggre_year_list:
        curr_yr = curr_st_name+year_name+"/"
        aggre_file_list = os.listdir(curr_yr)
        
        for file_name in aggre_file_list:
            curr_file = curr_yr+file_name
            data = open(curr_file,"r")
            
            Agg_trans = json.load(data)
            
            for i in Agg_trans["data"]["transactionData"]:
                name = i["name"]
                count = i["paymentInstruments"][0]["count"]
                amount = i["paymentInstruments"][0]["amount"]
                columns1["States"].append(state_name)
                columns1["Year"].append(int(year_name))
                columns1["Quarter"].append(int(file_name.strip(".json")))
                columns1["Transaction_type"].append(name)
                columns1["Transaction_count"].append(count)
                columns1["Transaction_amount"].append(amount)

                
aggregated_transaction = pd.DataFrame(columns1)                
aggregated_transaction["States"] = aggregated_transaction["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggregated_transaction["States"] = aggregated_transaction["States"].str.replace("-"," ")
aggregated_transaction["States"] = aggregated_transaction["States"].str.title()
aggregated_transaction['States'] = aggregated_transaction['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")

    
```

```python
#aggregated_user.
path2 = "C:/Users/mahes/OneDrive/Desktop/phonepe/pulse/data/aggregated/user/country/india/state/"
aggre_user_state_list = os.listdir(path2)

columns2 = {"States":[], "Year":[], "Quarter":[], "User_brand":[], "User_count":[], "User_percentage":[]}
for state_name in aggre_user_state_list:
    curr_st_name = path2+state_name+"/"
    year_list = os.listdir(curr_st_name)
    
    for year_name in year_list:
        curr_file_name = curr_st_name+year_name+"/"
        file_list = os.listdir(curr_file_name)
        
        for file_name in file_list:
            curr_doc = curr_file_name+file_name
            data = open(curr_doc,"r")
            
            agg_user = json.load(data)
            
            try:
                for i in agg_user["data"]["usersByDevice"]:
                    brand = i["brand"]
                    count = i["count"]
                    percentage = i["percentage"]
                    columns2["States"].append(state_name)
                    columns2["Year"].append(int(year_name))
                    columns2["Quarter"].append(int(file_name.strip(".json")))
                    columns2["User_brand"].append(brand)
                    columns2["User_count"].append(count)
                    columns2["User_percentage"].append(percentage)
            except:
                pass

aggregated_user = pd.DataFrame(columns2)
aggregated_user["States"] = aggregated_user["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggregated_user["States"] = aggregated_user["States"].str.replace("-"," ")
aggregated_user["States"] = aggregated_user["States"].str.title()
aggregated_user['States'] = aggregated_user['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
```
